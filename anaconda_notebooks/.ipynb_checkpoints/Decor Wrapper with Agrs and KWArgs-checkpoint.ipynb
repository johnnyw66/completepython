```python
import sys


class BaseDebug():
        @staticmethod
        def who_am_i(frame_index=1):
                # frame '1' is the 'calling frame'
                return sys._getframe(frame_index).f_code.co_name

        @staticmethod
        def where_am_i(frame_index=1):
                return sys._getframe(frame_index).f_lineno

        @staticmethod
        def which_file(frame_index= 1):
                return sys._getframe(frame_index).f_code.co_filename


class decorator_with_arguments(BaseDebug):

    def __init__(self, *args, **kwargs):
        """
        If there are decorator arguments, the function
        to be decorated is not passed to the constructor!
        """
        print("Inside __init__()")
        self.decor_args = args
        self.decor_kwargs = kwargs
        

    def __call__(self, f):
        """
        If there are decorator arguments, __call__() is only called
        once, as part of the decoration process! You can only give
        it a single argument, which is the function object.
        """
        #print(f"Inside __call__() LINE NUM {BaseDebug.where_am_i()}")
        
        def wrapped_f(*args):
            line_num = BaseDebug.where_am_i(2)
            source_file = BaseDebug.which_file(2)
            
            print(f"CALLED FROM LINE NUM {line_num} FILE {source_file}")
            print("Inside wrapped_f()")
            print("Decorator arguments:", self.decor_args[0], self.decor_args[1], self.decor_args[2],self.decor_kwargs['zippy'])
            print("Decorator arguments types:", type(self.decor_args[0]), type(self.decor_args[1]), type(self.decor_args[2])) 
            print("Decorator kwarguments types:", type(self.decor_kwargs['zippy']))
            print("Function Arguments ") 
            res = f(*args)
            print(f"After f(*args)")
            return res 
        return wrapped_f

@decorator_with_arguments("hello", "world", 42, zippy = True,level=0)
def sayHello(a1, a2, a3, a4):
    print('EXECUTING ... sayHello arguments:', a1, a2, a3, a4)
    return 666 

print("After decoration")

print("Preparing to call sayHello()")
res = sayHello("function_arg1", "function_arg2", "function_arg3", "function_arg4")
print(f"after first sayHello() call {res}")
sayHello("a", "different", "set of", "arguments")
print("THIS LINE IS ",BaseDebug.where_am_i())
```
